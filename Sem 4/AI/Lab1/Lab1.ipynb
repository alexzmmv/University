```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(44)

t = np.arange(0, 30, 0.01)
nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2

# Two signals with a coherent part at 10 Hz and a random part
s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, NFFT=256, Fs=1. / 0.01)
axs[1].set_ylabel('Coherence')

plt.show()
```

# Task 7


*Monte Carlo Simualation*

Consider a large pipe with a radius R=30cm that has the opening pointing straight up, as in the figure. Around this pipe is a rectangular wall with an edge l=1m. Balls of radius r=1cm fall down uniformly from above inside the square yard. Compute, using a Monte Carlo simulation, the probability that a ball falls inside the pipe. Compare the results with the mathematically deduced probability.



Considering the best position for the cilinder to be at the center of the square, the probability of a ball falling inside the pipe is the ratio of the area of the pipe to the area of the square.

```python
# Random positions in a 1m x 1m area
def computeprob(num_of_balls):
    ball_drop_location = torch.distributions.Uniform(0, 100).sample((num_of_balls, 2))
    ball_in_circle= ball_drop_location[torch.sqrt((ball_drop_location[:,0]-50)**2 + (ball_drop_location[:,1]-50)**2) < 30]
    return len(ball_in_circle)/num_of_balls

print(computeprob(90000))
```

Now mathematically, the probability of a ball falling inside the pipe is the ratio of the area of the pipe to the area of the square.

The area of the pipe is given by the formula: $\pi r^2$ and the area of the square is given by the formula: $l^2$
So the probability of a ball falling inside the pipe is given by: $\frac{\pi r^2}{l^2}$ where $r=30cm$ and $l=100cm$
